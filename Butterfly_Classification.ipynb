```python
modelPath = '/content/drive/My Drive/Butterfly Classification/Butterfly Segmentation/runs/segment/train/weights/best.pt'
segmentationModel = YOLO(modelPath)
```

```python
imageTestPath = '/content/drive/My Drive/Butterfly Classification/Butterfly Segmentation/SegmentationTest'
imageTest = [file for file in os.listdir(imageTestPath)]
```

```python
def imageSegmentation(image):
  results = segmentationModel(image)

  if results and results[0].masks and len(results[0].masks.data) > 0:
    mask = results[0].masks.data[0].numpy() * 255
    mask = cv2.resize(mask, (image.shape[1], image.shape[0]))
    mask = mask.astype(np.uint8)
    mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
    maskedImage = cv2.bitwise_and(image, mask)
    maskedImage = cv2.resize(maskedImage, (image.shape[1], image.shape[0]))
    return maskedImage
  else:
    return image
```

```python
fig, axes = plt.subplots(3, 5, figsize=(18, 12))
fig.suptitle('Butterfly Segmentation', fontsize=16)

for i, imageName in enumerate(imageTest):
  imagePath = os.path.join(imageTestPath, imageName)
  image = cv2.imread(imagePath)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

  results = segmentationModel.predict(image, verbose=False)
  predictedImage = results[0].plot()
  segmentedImage = imageSegmentation(image)

  axes[0, i].imshow(image)
  axes[1, i].imshow(predictedImage)
  axes[2, i].imshow(segmentedImage)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
dataLabel = pd.read_csv("drive/MyDrive/Butterfly Classification/Butterfly Classification/Training_set.csv")
```

```python
train, validation = train_test_split(dataLabel, test_size=0.2, random_state=14, stratify=dataLabel['label'])
```

```python
trainDataPreprocessing = ImageDataGenerator(
    rescale = 1/255,
    rotation_range = 20,
    width_shift_range = 0.1,
    height_shift_range = 0.1,
    zoom_range = 0.15,
    horizontal_flip = True,
    vertical_flip = True,
    fill_mode = 'nearest'
)

validationDataPreprocessing = ImageDataGenerator(
    rescale = 1/255
)
```

```python
imagePath = 'train/segmentedTrainImages'
```

```python
trainDataIterator = trainDataPreprocessing.flow_from_dataframe(
    dataframe = train,
    directory = imagePath,
    target_size = (256,256),
    x_col = 'filename',
    y_col = 'label',
    batch_size = 32,
    class_mode = 'categorical',
    shuffle=True
)

validationDataIterator = validationDataPreprocessing.flow_from_dataframe(
    dataframe=validation,
    directory=imagePath,
    target_size=(256, 256),
    x_col='filename',
    y_col='label',
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)
```

```python
base = tf.keras.applications.ConvNeXtLarge(weights="imagenet", include_top=False, input_shape=(256, 256, 3))
```

```python
base.trainable = False
```

```python
x = tf.keras.layers.GlobalAveragePooling2D()(base.output)
x = tf.keras.layers.Dense(512, activation='relu')(x)
x = tf.keras.layers.Dropout(0.1)(x)
output = tf.keras.layers.Dense(75, activation='softmax')(x)
```

```python
model = tf.keras.models.Model(inputs=base.input, outputs=output)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(), loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=5,
    restore_best_weights=True,
    verbose=1
)
```

```python
history = model.fit(trainDataIterator, epochs=50, validation_data=validationDataIterator, callbacks=[early_stopping])
```

```python
model.save('/content/drive/MyDrive/Butterfly Classification/Butterfly Classification/ButterflyClassificationModel.keras')
model.save('/content/drive/MyDrive/Butterfly Classification/Butterfly Classification/ButterflyClassificationModel.h5')
```

```python
fig = plt.figure()
plt.plot(history.history['accuracy'], color='teal', label='accuracy')
plt.plot(history.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(history.history['loss'], color='teal', label='loss')
plt.plot(history.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```
