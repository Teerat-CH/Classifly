```python
import os
import cv2
import numpy as np
from ultralytics import YOLO
from google.colab import drive
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
model = YOLO('yolov8s-seg.pt')
model.train(data='/content/drive/My Drive/Butterfly Classification/Butterfly Segmentation/config.yaml', epochs=15)
```

```python
os.chdir('/content')
```

```python
modelPath = '/content/drive/My Drive/Butterfly Classification/Butterfly Segmentation/runs/segment/train/weights/best.pt'
segmentationModel = YOLO(modelPath)
```

```python
imageTestPath = '/content/drive/My Drive/Butterfly Classification/Butterfly Segmentation/SegmentationTest'
imageTest = [file for file in os.listdir(imageTestPath)]
```

```python
fig, axes = plt.subplots(2, 5, figsize=(18, 8))
fig.suptitle('Original vs Predicted Images', fontsize=16)

for i, imageName in enumerate(imageTest):
  imagePath = os.path.join(imageTestPath, imageName)
  image = cv2.imread(imagePath)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

  results = segmentationModel.predict(image, verbose=False)
  predicted_image = results[0].plot()

  axes[0, i].imshow(image)
  axes[1, i].imshow(predicted_image)

plt.tight_layout()
plt.show()
```
